```python
import openai
import os, dotenv

dotenv.load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

openai.api_key = OPENAI_API_KEY
```

```python
import time
overall_start_time = time.time()
client = openai.OpenAI()

# Create a new conversation
start_time = time.time()
conversation = openai.conversations.create()
end_time = time.time()
print(f"Time taken to create a new conversation: {end_time - start_time} seconds")
print(conversation)
print("-"*10)

# Check the conversation. It's empty for now.
start_time = time.time()
items = client.conversations.items.list(conversation.id, limit=10)
end_time = time.time()
print(f"Time taken to check the conversation: {end_time - start_time} seconds")
print(items)
print("-"*10)

# Send a message to the conversation and get an AI response
start_time = time.time()
response = client.responses.create(
    model="gpt-4.1-mini",
    conversation=conversation.id,
    input="Hello. I'm Min Khant. Nice to meet you."
)
end_time = time.time()
print(f"Time taken to send a message and get an AI response: {end_time - start_time} seconds")
print(response)
print("-"*10)

# Parse the response to get the AI's response
start_time = time.time()
ai_response = response.output[0].content[0].text
end_time = time.time()
print(f"Time taken to parse the response: {end_time - start_time} seconds")
print(ai_response)
print("-"*10)

# Check the conversation again. It should now have the user's message and the AI's response.
start_time = time.time()
items = client.conversations.items.list(conversation.id, limit=10)
end_time = time.time()
print(f"Time taken to check the conversation again: {end_time - start_time} seconds")
print(items)
print("-"*10)

overall_end_time = time.time()
print(f"Total time taken: {overall_end_time - overall_start_time} seconds")
```
